```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# File path (adjust this to your actual file location)
file_path = "potential.xvg"

# Lists to store the data
time = []
potential = []

# Read the file, skipping header lines
with open(file_path, 'r') as file:
    for line in file:
        # Skip lines starting with '#' or '@'
        if line.startswith('#') or line.startswith('@'):
            continue
        # Split the line into columns and convert to float
        columns = line.split()
        time.append(float(columns[0]))
        potential.append(float(columns[1]))

# Convert lists to numpy arrays for easier handling
time = np.array(time)
potential = np.array(potential)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(time, potential, label="Potential", color="blue")
plt.xlabel("Energy Minimization step")
plt.ylabel("Potential Energy (kJ/mol)")
plt.title("Energy Minimization")
plt.legend()
plt.grid(True)

# Adjust layout to match the view settings from the file (optional)
plt.subplots_adjust(left=0.15, bottom=0.15, right=0.75, top=0.85)

# Show the plot
plt.show()
```

```python
# File path (adjust this to your actual file location)
file_path = "temperature.xvg"

# Lists to store the data
time = []
temperature = []

# Read the file, skipping header lines
with open(file_path, 'r') as file:
    for line in file:
        # Skip lines starting with '#' or '@'
        if line.startswith('#') or line.startswith('@'):
            continue
        # Split the line into columns and convert to float
        columns = line.split()
        time.append(float(columns[0]))
        temperature.append(float(columns[1]))

# Convert lists to numpy arrays for easier handling
time = np.array(time)
temperature = np.array(temperature)

# Calculate total time and time interval
total_time = time[-1] - time[0]
num_points = len(time)
time_interval = total_time / (num_points - 1) if num_points > 1 else 0
print(f"Total time: {total_time:.2f} ps, Time interval: {time_interval:.4f} ps, Points: {num_points}")

# Determine window size for running average based on total time
window_time = total_time / 10  

# Calculate window size in points
window_size = int(window_time / time_interval) if time_interval > 0 else num_points
if window_size > num_points:
    window_size = num_points  # Cap at total points if window exceeds data
elif window_size < 1:
    window_size = 1  # Minimum window size of 1 point
print(f"Running average window: {window_time} ps ({window_size} points)")

# Calculate the 10 ps running average
running_avg = np.convolve(temperature, np.ones(window_size)/window_size, mode='valid')
#print(running_avg)
time_avg = time[window_size-1:]  # Adjust time to match running average

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(time, temperature, label="Temperature", color="blue")
plt.plot(time_avg, running_avg, label=f"{window_time} ps Running Average", color="red", linewidth=2)  # Running average
plt.xlabel("Time (ps)")
plt.ylabel("Temperature (K)")
plt.title("NVT Equilibration Temperature")
plt.legend()
plt.grid(True)

# Adjust layout to match the view settings from the file (optional)
plt.subplots_adjust(left=0.15, bottom=0.15, right=0.75, top=0.85)

# Show the plot
plt.show()
```
